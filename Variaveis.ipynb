```python
#Principais tipos de variáveis

#Inteiros (int): Números inteiros, sem parte decimal.
int = 23
print("Inteiro: ", int)

#Ponto Flutuante (float): Números reais. que têm uma parte decimal.
float = 23.1
print("Flutuante: ", float)

#Complexos (complex): Números complexos, que têm uma parte real e uma parte imaginária.
complex = 23 + 4j
print("Complexo: ", complex)

#Strings (str): sequência de caracteres.
str = "Olá, Mundo!"
print("Texto: ", str)

"""
Os termos "mutável" e "imutável" referem-se à capacidade de alterar o conteúdo de um objeto depois que ele foi criado

Mutável: Significa que o conteúdo do objeto pode ser alterado após a sua criação As listas em Py são mutáveis, o que significa que você pode
modificar seus elementos.

Imutável: significa que o conteúdo do objeto não ode ser alterado após a sua criação. As tuplas em Py são imutáveis, o que significa que, uma vez
criada, você não pode mais modificar seus elementos.
"""

#Listas:
list = [1, 2, 3]
print("Lista", list)

#Tuplas (tuple): Uma coleção ordenada e imutável.
tuple = (1, 2, 3, 4, 5)
print("Tupla:", tuple)

#Conjuntos (set): uma coleção não ordenada de itens únicos.
set = {1, 2, 3}
print("Conjunto:", set)

#Dicionários (dict): Uma coleção não ordenada de pares chave-valor.
dict = {'key': 'val'}
print("Dicionário:", dict)

#Booleanos (bool): valores verdadeiros ou falsos.
bool = False
print('Booleano:', bool)

#Nonetype (none): Um tipo especial que representa a ausência de valor.
nenhum = None
print('NoneType:', nenhum)
```

```python
#Podemos obter o tipo de uma variável usando a função type

"""
nome = str("Ana Paula"): A string é passada para a função str(), que retorna a mesma string. A função str() é desnecessária neste caso, pois "Ana Paula"
já é uma String. A Var nome é atribuída a essa string.

idade = int(31): o número inteiro 31 é passado para a função int(), que retorna o mesmo número. Novamente, a função int() é desnecessária aqui, pois 31
já é um inteiro. A variável idade é atribuída a esse valor inteiro.

print(type(nome)): a função 'type()' retorna o tipo da variável nome, que é <class 'str'>, já que nome é uma str. Esse valor é então impresso, resultan-
do na saída:
"""

nome = "Ana Paula"
idade = 31

print(type(nome))
print(type(idade))

#As variáveis diferenciam letras maiúsculas de minúsculas

i = 30
I = "Ana"
print(i)
```

```python
"""
"Slicing" (fatiamento) em python para extrair partes de uma string
    frase = "Olá, mundo!" >> A linha atribui a string "Olá, mundo!" à variável frase.

    parte = frase[4:8] >> Isto fatia a string da posição 4 até a posição 8 (exclusiva). Em python, a index começa em 0, então a posição 4 é o quinto caractere da string, que é um espaço, e a posição 8 é o nono caractere, que é a letra 'd'. Então, essa fatia inclui os caracteres nas posições 4, 5, 6 e 7, que são " mun". O valor de parte será a string " mun".
    
    print(parte) >> imprime o valor de parte, que é " mun".

    primeiros = frase[:5] >> isso fatia s string da posição inicial (inclusa) até a posição 5 (exclusiva). Como não há um valor antes dos dois pontos, ele assume o início da string. Portanto, essa fatia inclui os caracteres nas posições 0, 1, 2, 3, e 4, que são "Olá,". O val de primeiros será a string "Olá,".

print(ultimos): imprime o valor de ultimos, que é "mundo!"
"""
# Slice
frase = "Olá, mundo!"

#obtendo uma parte da string usando slice
parte = frase[4:8]
print(parte) #saída: mun

#Obtendo os primeiros 5 caracteres da string
primeiros = frase[:5]
print(primeiros) #saída: Olá,

#Obtendo os últimos 6 caracteres da string
ultimos = frase[-6:]
print(ultimos) #saída: mundo!



frase = "O módulo de python é muito legal!"
if "python" in frase:
    print("Existe, sim, a palavra python")
    
#strip >> usamos para remover espaços em branco do início e fim da frase.
frase = "          O módulo de python é muito legal!          "
print(frase)
print(frase.strip()) #frase "strippada"

#split >> divide uma string em palavra ou partes menores. Ele utiliza os espaços em branco para determinar a separação das palavras.
#split, join e strip são métodos muito úteis da str

frase = "Olá, como vai você?"
palavra = frase.split()
print(palavra)

#Join >> faz a junção das palavras, eliminando os espaçamentos.
palavra = ['Olá,', 'como', 'vai', 'você?']
frase = ' '.join(palavra)
print(frase)
```
